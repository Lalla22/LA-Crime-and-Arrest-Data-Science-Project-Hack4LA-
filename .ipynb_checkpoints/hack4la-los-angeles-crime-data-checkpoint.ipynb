```python
# Histogram with KDE for 'Victim Age' column
plt.figure(figsize=(10, 6))
sns.histplot(df_crime['Victim_Age'], kde=True, bins=30, color='blue')
plt.title('Distribution of Age')
plt.xlabel('Victim Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram with KDE for 'Victim Sex' column
plt.figure(figsize=(10, 6))
sns.histplot(df_crime['Victim_Sex'], kde=True, bins=30, color='blue')
plt.title('Distribution of Age')
plt.xlabel('Victim Sex')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram with KDE for 'Victim Age' column
plt.figure(figsize=(10, 6))
sns.histplot(df_crime['Victim_Age'], kde=True, bins=30, color='blue')
plt.title('Distribution of Age')
plt.xlabel('Victim Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot for top 10 'Crime_Code_Description'
top_charge_groups = df_crime['Crime_Code_Description'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_charge_groups.values, y=top_charge_groups.index, palette='viridis')
plt.title('Top 10 Charge Groups')
plt.xlabel('Count')
plt.ylabel('Crime Group Description')
plt.show()
```

```python
# Calculate correlation matrix
correlation_matrix = df_crime.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
#Histogram 
sns.distplot(df_crime['Victim_Age'])
```

```python
# Histogram and KDE plot of time crime happened
plt.figure(figsize=(10, 6))
sns.histplot(df_crime['Time_Occurred'], kde=True, color='blue')
plt.title('Distribution of Time')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.show()
```
